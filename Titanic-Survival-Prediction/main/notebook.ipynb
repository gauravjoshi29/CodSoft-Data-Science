```python
alone_mapping = {label: idx for idx, label in enumerate(np.unique(df['alone']))}
adult_male_mapping = {label: idx for idx, label in enumerate(np.unique(df['adult_male']))}
sex_mapping = {label: idx for idx, label in enumerate(np.unique(df['sex']))}
class_mapping = {label: idx for idx, label in enumerate(np.unique(df['class']))}
```

```python
alone_mapping
```

```python
adult_male_mapping
```

```python
sex_mapping
```

```python
class_mapping
```

```python
df['alone'] = df['alone'].map(alone_mapping)
df['adult_male'] = df['adult_male'].map(adult_male_mapping)
df['sex'] = df['sex'].map(sex_mapping)
df['class'] = df['class'].map(class_mapping)
```

```python
df.head()
```

```python
x = df.iloc[:, df.columns != 'survived']
y = df.iloc[:, df.columns == 'survived']
```

```python
x
```

```python
y
```

```python
#heatmap
plt.figure(figsize = (8,4))
sns.heatmap(x.corr(), annot = True)
```

```python
#train test split
from sklearn.model_selection import train_test_split, GridSearchCV
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
log_classifier = LogisticRegression()
log_classifier
```

```python
import time
np.random.seed(42)
start = time.time()
parameter = {'penalty': ['l1','l2','elasticnet'],
             'C': [1,2,3,4,5,6,10,20,30,40,50],
             'max_iter': [100,200,300]}

cv_logr = GridSearchCV(log_classifier, param_grid = parameter, scoring = 'accuracy', cv = 5)

cv_logr.fit(x_train, y_train)
print("Best Parameters:",cv_logr.best_params_)
end = time.time()
print("Time Taken:%0.2f"%(end-start))
```

```python
log_classifier.set_params(C = 4, max_iter = 100, penalty = 'l2')
```

```python
log_classifier.fit(x_train, y_train)
```

```python
pred = log_classifier.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
score = accuracy_score(pred, y_test)
score
```

```python
#classification report
print(classification_report(pred, y_test))
```

```python
#confusion matrix
cmat = confusion_matrix(y_test, pred)
cmat
```

```python
#in heatmap
sns.heatmap(cmat, annot = True)
```
